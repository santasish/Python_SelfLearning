#### There are 2 types of Data-types in Pandas library.
1. Series: for single column/ single row          ........  [list]
2. DataFrame: for complete tabular structure      ........ DataFrames are basically combination of multiple Series

Series: It is similar to a list! Heterogenous datatypes,indexed/ordered. Pandas library recognizes it as a series when a single row / column is being operated on. This single row/column (series) is like a single List.

DataFrame: Combination of multiple columns(singular: Series) vertically or multiple rows(singular: Series) horizontally. They provide for the complete tabular structure.

.iloc and .loc giving the same result!

iloc always goes for the default inbuilt internal indexes.[index location or integer location]

loc goes after named indexes. or direct location!

Since no rows were deleted from the dataset permanently, both the indexes are same. If a row was deleted, the internal indexing mechanism of pandas would rearrange the indexing as 0,1,2... which would then differ from the named indexes which has been named by the user!

 If we delete the first record, iloc would become 0 for second record  but loc would remain as 1.
