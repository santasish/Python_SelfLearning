# NESTED LOO[P]

```python
#_STUDY_Example1:

n = 5                                         # Think of a 5x5 Matrix  ['i' bulids rows & 'j' builds columns]

for i in range(0,n):                          # GIVES A LIST ---> [0,1,2,3,4] ---> initial value of i = 0 (Row Number)
                                 
                                #_movement of i--> [i=0,i+1=1]; [i=1,i+1=2]; [i=2,i+1=3]; [i=3,i+1=4]; [i=4,i+1=5]
    for j in range(0,i+1):       # nested for loop produces range --> [j=0,i=1]; [j=1,i=2]; [j=2,i=3]; [j=3,i=4]
                                                           
        
        print("santa", end = " ") # print 1, end: prints in the next line after putting space; next iterations [print 2 the next time, put space; print 3 times....]  
    
    print("\r")                   # change line when the main for loop finishes each iteration or changes each row!


    #[DISCLAIMER]:

# \n o kora jeto --> \n is for a new line.. jota icha new line kora jabe
# \r is for the next line... shob shomoy immediate next line ei jabe... 
# hash out 'end' & check ki hoy ?!
# 'i' bulids rows & 'j' builds columns

# """""""#_OUTPUT_
# (0)                      i=1     j=0
# (0,1)                    i=2     j=0,j=1
# (0,1,2)                  i=3     j=0,j=1,j=2
# (0,1,2,3)                i=4     j=0,j=1,j=2,j=3
# (0,1,2,3,4)              i=5     j=0,j=1,j=2,j=3,j=4""""""""
```

```python
# using FOR-eLSE

word = input()

for i in range(len(word)):
    if word[i] == word[len(word)-1-i]:                     # mapping the indexes
        continue
    else:
        print(f"{word} is NOT PALINDROME")
else:
    print(f"{word} is a PALINDROME")
```

```python
# Ex2: using FOR-eLSE

word = input()

for i in range(len(word)):
    if word[i] != word[len(word)-i-1]:
        print(f"{word} is not a Palinmdrome")
        break
else:
    print(f"{word} is a palindrome")
```

```python
# using WHILE loop

s = input()                              # user input

f = str()                                # creating a new string

i= len(s)-1                              # initializing the iterator from the reverse

while i>=0:
    f = f + s[i]                         # appending the new string,f
    i = i - 1                            # iterating the while loop in reverse
    
if s == f :
    print(f"{s} is palindrome")          # if-else to compare str and it's reverse
else:
    print(f"{s} is NOT PALINDROME")
```

```python
#code to find the max values in the dictionary d_1

list_n= list()

for i in d_1.values():
    for j in i.values():
        list_n.append(j)
    print(max(list_n))
    
```

```python
#Approach 2

for i in d_1.values():            #........[14,10,4]......[45,34,1]
    mx = 0                        #.... flag created
    for j in i.values():          #........[14,10,4] ; then......[45,34,1]
        if mx<j:
            mx = j                # 1st iteration--> mx=0 <j, mx=j --> mx=14....
    print(mx)
    
```

```python
#Approach 3 (a)

for i in d_1.values():
    print(max(i.values()))
    
```

```python
#Approach 3(b)

for i in d_1.keys():
    print(i)
    print(max(d_1[i].values()))
    
```
